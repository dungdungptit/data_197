````python
# df1 = columns reactions
df1 = df_pages.loc[:, ["reactions"]]
df1.head()
````

````python
# eval() to convert string to list
df1 = df1.applymap(lambda x: eval(x))
df1 = df1.reactions.apply(pd.Series)

df1.head()
````

````python
df1 = df1.fillna(0)
df1.head()
````

````python
# reactions = {'th√≠ch': 306, 'y√™u th√≠ch': 37, 'haha': 64, 'wow': 2, 'th∆∞∆°ng th∆∞∆°ng': 4, 'bu·ªìn': 5} to columns
df_pages = pd.concat([df_pages, df1], axis=1)
df_pages = df_pages.drop(columns=["reactions"])
df_pages.head()
````

# x·ª≠ l√≠ comments_full

````python
import datetime
df2 = df_pages.loc[:, ["comments_full"]]
df2["comments_full"] = df2["comments_full"].apply(lambda x: eval(x))
df2
````

````python
# check type of comments_full
print(df2["comments_full"][1][0].keys())
# create new column commenter_name
df2["commenter_name"] = np.nan
for i in range(len(df2)):
    comment_text = ""
    commenter_name = ""
    # create new column commenter_name
    if type(df2["comments_full"][i]) != list:
        print(i)
    if len(df2["comments_full"][i]):
        for j in range(len(df2["comments_full"][i])):
            comment_text += df2["comments_full"][i][j]["comment_text"] + " "
            # print(df2["comments_full"][i][j]["commenter_name"])
            commenter_name += df2["comments_full"][i][j]["commenter_name"] + ", "
        
        # print replies
        if len(df2["comments_full"][i][j]["replies"]):
            for k in range(len(df2["comments_full"][i][j]["replies"])):
                comment_text += df2["comments_full"][i][j]["replies"][k]["comment_text"] + " "
                commenter_name += df2["comments_full"][i][j]["replies"][k]["commenter_name"] + ", "
    else:
        comment_text = "No comment"

    df2["comments_full"][i] = comment_text
    df2["commenter_name"][i] = commenter_name
    if df2["commenter_name"][i] == "":
        df2["commenter_name"][i] = "No comment"

df2.head()    
````

````python
df_pages = df_pages.drop(columns=["comments_full"])
df_pages = pd.concat([df_pages, df2], axis=1)
df_pages.head()
````

````python
df_pages.to_csv("data_xuli_16112023.csv", index=False, encoding="utf-8-sig")
````
